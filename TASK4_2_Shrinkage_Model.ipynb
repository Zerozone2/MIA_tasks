```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import *
```

```python
df = pd.read_csv('shrink_ray_dataset.csv')
cpdf = df.copy()
cpdf.head()
```

```python
y = np.array(cpdf['Power'])
X = np.array(cpdf["Shrinkage"]).reshape(-1, 1)
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
 
model = LinearRegression()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
result1 = model.predict(np.array([[85]]))
print(result1[0])
result2 = model.intercept_ + model.coef_ * 85
print(result2[0])
```

```python
sns.lmplot(x ="Power", y ="Shrinkage", data = cpdf )
plt.scatter(y_test, X_test, color = 'k', label = "sample")
plt.legend()
plt.show()
```
